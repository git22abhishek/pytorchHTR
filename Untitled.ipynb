## Resize

```python

image = io.imread('/mnt/d/Machine-Learning/Datasets/bentham/uncompressed/BenthamDatasetR0-GT/Images/Lines/002_080_001_03_01.png')

seg_mask = image[:,:,3]
rgb_image = image[:, :, :-1]

gray = cv2.cvtColor(rgb_image, cv2.COLOR_RGB2GRAY) # skimage.color.rbg2gray does not preserve range 


segmented_image = (gray * (seg_mask == 255 * 1)) + ((seg_mask == 0) * 195).astype(np.uint8)
deslanted_image = deslant(segmented_image, bg_color=255).img

_, binary = cv2.threshold(deslanted_image, 0, 255, cv2.THRESH_BINARY+cv2.THRESH_OTSU)

al_resize = A.augmentations.geometric.Resize(height=128, width=1024, always_apply=True)
albu_resized = al_resize(image=binary)['image']


resize = (128, 1024)
h, w = deslanted_image.shape[:2]
fx = w / resize[1]
fy = h / resize[0]
f = max(fx, fy)
new_size = (max(min(resize[0], int(h / f)), 1), max(min(resize[1], int(w / f)), 1))
sk_resized = skimage.transform.resize(binary, new_size, preserve_range=True, mode='constant', cval=255, anti_aliasing=True)


f, axes = plt.subplots(2, 1, figsize=(20, 10))

# we use `axes.ravel()` to turn these into a list
axes = axes.ravel()

for ax in axes:
    ax.axis('off') # switch off the axis (lose the x and y axis tick marks on the edges)

axes[0].imshow(albu_resized)
axes[0].set_title('Aspect Ratio not preserved')

axes[1].imshow(sk_resized)
axes[1].set_title('Aspect Ratio preserved')

print(albu_resized.shape)
print(sk_resized.shape)
```
